Sistema básico do salão

```python
from datetime import date
from IPython.display import display, clear_output
import ipywidgets as widgets

valor_hidrat = []
valor_prog = []
valor_mani = []
valor_pedi = []
valor_relaxante = []
valor_fisio = []

output = widgets.Output()

dia = date.today()
data_br = dia.strftime("%d/%m/%Y")

def reiniciar():
    clear_output()
    main()

def main():
    print(f"Dia: {data_br}\n")

    dropdown = widgets.Dropdown(
        options=[("Cabelo", 1), ("Manicure/Pedicure", 2), ("Massagem", 3), ("Encerrar dia", 4)],
        value=None,
        description="Serviço:",
    )
    botao = widgets.Button(description="Confirmar")

    def ao_clicar(b):
        with output:
            clear_output()

        if dropdown.value is None:
            with output:
                print("Por favor, selecione um serviço.")
            return

        if dropdown.value == 1:
            with output:
                print("Você escolheu: Cabelo")

            tipo = widgets.Dropdown(
                options=["Hidratação", "Progressiva"],
                description="Procedimento:",
                value=None
            )
            botao_tipo = widgets.Button(description="Confirmar")

            def confirmar(b2):
                if tipo.value is None:
                    with output:
                        clear_output()
                        print("Selecione um procedimento.")
                    return

                produto = widgets.Text(description="Produto:")
                cliente = widgets.Text(description="Cliente:")
                valor = widgets.IntText(description="Valor (R$):")
                botao_final = widgets.Button(description="Salvar")

                def salvar(b3):
                    with output:
                        clear_output()
                        print(f"Cabelo - {tipo.value}")
                        print(f"Produto: {produto.value}")
                        print(f"Cliente: {cliente.value}")
                        print(f"Valor: R$ {valor.value:.2f}")
                        if tipo.value == "Hidratação":
                            valor_hidrat.append(valor.value)
                        else:
                            valor_prog.append(valor.value)
                    reiniciar()

                botao_final.on_click(salvar)
                display(produto, cliente, valor, botao_final)

            botao_tipo.on_click(confirmar)
            display(tipo, botao_tipo)

        elif dropdown.value == 2:
            with output:
                print("Você escolheu: Manicure/Pedicure")

            tipo = widgets.Dropdown(
                options=[("Manicure"), ("Pedicure")],
                description="Procedimento:",
                value=None
            )
            botao_tipo = widgets.Button(description="Confirmar")

            def confirmar(b2):
                if tipo.value is None:
                    with output:
                        clear_output()
                        print("Selecione um procedimento.")
                    return

                esmalte = widgets.Text(description="Esmalte:")
                cliente = widgets.Text(description="Cliente:")
                valor = widgets.IntText(description="Valor (R$):")
                botao_final = widgets.Button(description="Salvar")

                def salvar(b3):
                    with output:
                        clear_output()
                        print(f"{tipo.value}")
                        print(f"Esmalte: {esmalte.value}")
                        print(f"Cliente: {cliente.value}")
                        print(f"Valor: R$ {valor.value:.2f}")
                        if tipo.value == "Manicure":
                            valor_mani.append(valor.value)
                        else:
                            valor_pedi.append(valor.value)
                    reiniciar()

                botao_final.on_click(salvar)
                display(esmalte, cliente, valor, botao_final)

            botao_tipo.on_click(confirmar)
            display(tipo, botao_tipo)

        elif dropdown.value == 3:
            with output:
                print("Você escolheu: Massagem")

            tipo = widgets.Dropdown(
                options=[("Fisioterapia"), ("Relaxante")],
                description="Procedimento:",
                value=None
            )
            botao_tipo = widgets.Button(description="Confirmar")

            def confirmar(b2):
                if tipo.value is None:
                    with output:
                        clear_output()
                        print("Selecione um tipo de massagem.")
                    return

                duracao = widgets.IntText(description="Duração (min):")
                cliente = widgets.Text(description="Cliente:")
                valor = widgets.IntText(description="Valor (R$):")
                botao_final = widgets.Button(description="Salvar")

                def salvar(b3):
                    with output:
                        clear_output()
                        print(f"{tipo.value}")
                        print(f"Duração: {duracao.value} minutos")
                        print(f"Cliente: {cliente.value}")
                        print(f"Valor: R$ {valor.value:.2f}")
                        if tipo.value == "Fisioterapia":
                            valor_fisio.append(valor.value)
                        else:
                            valor_relaxante.append(valor.value)
                    reiniciar()

                botao_final.on_click(salvar)
                display(duracao, cliente, valor, botao_final)

            botao_tipo.on_click(confirmar)
            display(tipo, botao_tipo)

        elif dropdown.value == 4:
            with output:
                clear_output()
                print("Encerrando dia")
                print("Resumo do dia:")
                print(f"Hidratação: R$ {sum(valor_hidrat):.2f}")
                print(f"Progressiva: R$ {sum(valor_prog):.2f}")
                print(f"Manicure: R$ {sum(valor_mani):.2f}")
                print(f"Pedicure: R$ {sum(valor_pedi):.2f}")
                print(f"Relaxante: R$ {sum(valor_relaxante):.2f}")
                print(f"Fisioterapia: R$ {sum(valor_fisio):.2f}")
                total = sum(valor_hidrat) + sum(valor_prog) + sum(valor_mani) + sum(valor_pedi) + sum(valor_relaxante) + sum(valor_fisio)
                print(f"\nGanhos totais: R$ {total:.2f}")
                print("\nPrograma encerrado.")

    botao.on_click(ao_clicar)
    display(dropdown, botao, output)

reiniciar()
```
